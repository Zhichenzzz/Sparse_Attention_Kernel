```python
import pandas as pd
import matplotlib.pyplot as plt  

# 读取CSV文件
file_path = 'sparse_attn/test_topk_opt1/fused-attention-batch4-head32-d128-fwd-causal=True.csv'
df = pd.read_csv(file_path)

df = df.iloc[:50]  

# df['topk'] = df['topk'].apply(lambda x: float(x)/256)
df['Speedup'] = df['Flash-2'] / df['Triton [FP16]']
df['Ideal'] = df['topk'].apply(lambda x: 128/x if x != 0.0 else None)
pd.set_option('display.max_rows', 200)  # 设置显示所有行

df['ratio'] = df['Speedup'] / df['Ideal']
# 选择需要的列  
df_selected = df[['topk', 'Speedup', 'Ideal', 'ratio']]
  
# 创建一个包含两个子图的图表  
fig, ax1 = plt.subplots(figsize=(10, 6))  
  
# 绘制第一个子图 (主图)  
ax1.plot(df_selected['topk'], df_selected['Speedup'], label='Speedup', marker='o')  
ax1.plot(df_selected['topk'], df_selected['Ideal'], label='Ideal', marker='x')  
ax1.set_xlabel('topk')  
ax1.set_ylabel('Speedup')  
ax1.set_title('Speedup and Ideal via topk')  
ax1.legend()  
  
# 创建第二个子图 (嵌入小图)  
ax2 = fig.add_axes([0.6, 0.3, 0.25, 0.4])  # [left, bottom, width, height]  
ax2.plot(df_selected['topk'], df_selected['ratio'], label='Speedup/Ideal', marker='*', color='green')  
ax2.set_xlabel('topk')  
ax2.set_ylabel('Ratio')  
ax2.set_title('Speedup / Ideal')  
ax2.legend()  
  
# 显示图表  
plt.show()  
```

```python
import pandas as pd
import matplotlib.pyplot as plt  

# 读取CSV文件
file_path = '/home/v-zhizeng/Sparse_Atten_Kernel/sparse_attention/sparse_attn_16k/sparse-attention-batch4-head32-seq16384-causal=True.csv'
file_2 = '/home/v-zhizeng/Sparse_Atten_Kernel/predictor_kernel/mask_predictor/mask_predictor-batch4-hidden256-Seq16384-fwd.csv'
df1 = pd.read_csv(file_path)
df2 = pd.read_csv(file_2)

df1
df1['Ratio'] = df1['topk'].apply(lambda x: x/128 *100)
df2['Ratio'] = df2['topk'].apply(lambda x: x/128 *100)
df1['Speedup'] = df1['Flash-2'] / df1['Triton [FP16]']
df2['Speedup'] = df2['Torch-fp16'] / df2['Triton-fp16']

df3 = pd.merge(df1, df2, on='Ratio', how='inner')
df3
# df3 = df3[['Ratio', 'Speedup_x', 'Speedup_y']]
df3['sparse_attn'] = df3['Triton [FP16]'] + df3['Triton-fp16']
df3['full_attn'] = df3['Flash-2']
df3['Speedup'] = df3['full_attn'] / df3['sparse_attn']
df3 = df3[['Ratio', 'sparse_attn', 'full_attn','Speedup']]
df3= df3.iloc[1:]
# 创建图表
plt.figure(figsize=(10, 6))
plt.plot(df3['Ratio'], df3['Speedup'], marker='o')

# 添加标题和标签
# plt.title('Sparse Attention Speedup compared to Full Attention (flashattention-2)')
plt.xlabel('None-zero Ratio (%)', fontsize=14, fontweight='bold')
plt.ylabel('Speedup', fontsize=14, fontweight='bold')
# 加粗坐标轴
ax = plt.gca()  # 获取当前坐标轴
ax.spines['top'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)

# 加粗坐标轴刻度
ax.tick_params(axis='both', which='major', labelsize=12, width=1.5)

# 反转x轴
plt.xlim(0, 50)

plt.gca().invert_xaxis()
plt.axhline(y=1, color='r', linestyle='--')

# 显示图表
plt.grid(True)
plt.ylim(0, 20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt  

# 读取CSV文件
file_path = './fused_pooling_flash/pooling_flashattn_profiling/gpu_profile.csv'
df1 = pd.read_csv(file_path)

df1['N_CTX'] = df1['N_CTX'].apply(lambda x: x/1000)
df1['Ratio'] = df1['fapooling'] - df1['Triton']
max_ratio_index = df1['Ratio'].idxmax()
max_ratio_point = df1.iloc[max_ratio_index]
# # 创建图表
plt.figure(figsize=(10, 6))
plt.plot(df1['N_CTX'], df1['Triton'], marker='o', color='blue', label='Flash Attention')
plt.plot(df1['N_CTX'], df1['fapooling'], marker='o', color='green', label='Flash-Attn with pooling')
difference = max_ratio_point['fapooling'] - max_ratio_point['Triton']
plt.annotate(
    '', 
    xy=(max_ratio_point['N_CTX'], max_ratio_point['fapooling']), 
    xytext=(max_ratio_point['N_CTX'], max_ratio_point['Triton']),
    arrowprops=dict(arrowstyle='<->', color='red', linewidth=2)  # Adjust linewidth here
)
plt.text(max_ratio_point['N_CTX'], (max_ratio_point['fapooling'] + max_ratio_point['Triton']) / 2,
         f'1.35x ', fontsize=14, ha='right', va='top', color='red')

# # 添加标题和标签
# # plt.title('Sparse Attention Speedup compared to Full Attention (flashattention-2)')
plt.xlabel('Sequence Length (K)', fontsize=14, fontweight='bold')
plt.ylabel('GPU Memory Peak Usage (GB)', fontsize=14, fontweight='bold')
# 加粗坐标轴
ax = plt.gca()  # 获取当前坐标轴
ax.spines['top'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)

# 加粗坐标轴刻度
ax.tick_params(axis='both', which='major', labelsize=12, width=1.5)

plt.grid(True)
plt.legend()
plt.ylim(0, 80)
plt.show()
```
